# Credit Card Customer Characteristic

Reading data from csv file

```
data <- read.csv("BankChurners.csv")  
```

Attrition_Flag data

```
pie(table(data$Attrition_Flag), main = "Attrition Flag")
table(data$Attrition_Flag)
```

Customer Age Data Distribution

```
hist(data$Customer_Age, main = "Customer Age")
summary(data$Customer_Age)
#To Find mode
names(sort(-table(data$Customer_Age)))[1]
```

Customer Gender Data

```
pie(table(data$Gender), main = "Customer Gender")
table(data$Gender)
```

Card Category data

```
barplot(table(data$Card_Category), col = 4, main = "Card Category")
table(data$Card_Category)
```

Marital Status data

```
barplot(table(data$Marital_Status), main = "Marital Status", col = 3)
table(data$Marital_Status)
```

Education Level data distribution

```
pie(table(data$Education_Level), main = "Education Level")
table(data$Education_Level)
```

Credit Limit data distribution

```
hist(data$Credit_Limit, main = "Histogram of Credit Limit")
summary(data$Credit_Limit)
```

Correlation Test and Scatter Plot of Credit Limit and Avg Open To Buy

```
cor.test(data$Credit_Limit, data$Avg_Open_To_Buy)
```

```
plot(data$Credit_Limit, data$Avg_Open_To_Buy)
```
